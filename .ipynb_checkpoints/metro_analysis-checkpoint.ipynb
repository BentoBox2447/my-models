### Cleaning the Data
- break down the time columns into new columns ('full_date', 'time', 'year', 'month', 'date')
- rename time as time_stamp and leave it like that

```python
import pandas as pd
```

```python
data = pd.read_csv('data/hotcars.csv')
```

```python
data.head()
```

```python
# renaming time into time_stamp
data = data.rename(columns = {'time' : 'time_stamp'})
```

```python
# splitting up the time_stamp into a month day and year
data['year'] = pd.DatetimeIndex(data['time_stamp']).year
data['month'] = pd.DatetimeIndex(data['time_stamp']).month
data['day'] = pd.DatetimeIndex(data['time_stamp']).day
```

```python
# making a new full date column out of the time stamp
data['full_date']=[d.split('T')[0] for d in data.time_stamp]
```

```python
data.head()
```

```python
# visuilization for how many tweets they were per year
data.year.value_counts().plot.bar()
```

```python
data.shape
```

```python
# most complaints come from these cars (avoid at all costs!)
data.car_number.value_counts().head(10)
```

```python
# most hot car complaints on Red, Orange, and Yellow Lines
data.color.value_counts()
```

```python
data.color.value_counts().plot.bar()
```

## System reporting seeing how many breakages they are
- This data set looks at all breakages per any given day

```python
report = pd.read_csv('data/daily_system_reports.csv')
```

```python
print(report.shape)
print(report.dtypes)
report.head()
```

```python
report = report.rename(columns = {'day' : 'full_date'}) 
```

```python
report.head()
```

```python
# this will break down our full_date column and make 3 new columns ('year', 'month', 'day')
report['year']=[d.split('-')[0] for d in report.full_date]
report['month']=[d.split('-')[1] for d in report.full_date]
report['day']=[d.split('-')[2]for d in report.full_date]
```

```python
report.head()
```

```python
report.nlargest(10, ['escalators_num_breaks'])
```

```python
# this is splitting up the day format which is in yyyy/mm/dd currently to 3 columns year, month, and day
report['year']=[d.split('-')[0] for d in report.day]
report['month']=[d.split('-')[1] for d in report.day]
report['day']=[d.split('-')[2] for d in report.day]

report.head(5)
```
